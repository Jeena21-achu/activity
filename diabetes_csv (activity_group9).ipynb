# 1. create a countplot of outcome

```python
 import matplotlib.pyplot as plt 
import seaborn as sns  
import pandas as pd  
df= pd.read_csv("C:\\Users\\DELL\\Downloads\\diabetes_model.csv")
sns.countplot(x='Outcome',data=df)  
plt.show()
```

# 2. Analyse effect of age on outcome using boxplot

```python
import matplotlib.pyplot as plt
df= pd.read_csv("C:\\Users\\DELL\\Downloads\\diabetes_model.csv")  
sns.boxplot(x=df['Outcome'],y=df['Age'],data=df)
plt.title('effect of age on outcome')
plt.show()
```

# 3. check for chances of getting diabetes if it is common in family,using diabetes pedigree fn

```python
import seaborn as sns
import pandas as pd 
df= pd.read_csv("C:\\Users\\DELL\\Downloads\\diabetes_model.csv")
sns.violinplot(y=df['DiabetesPedigreeFunction'],x=df['Outcome'],data=df)
```

# 4 check for relationship with insulin and glucose and group datapoints by outcome

```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plot
df= pd.read_csv("C:\\Users\\DELL\\Downloads\\diabetes_model.csv")
plot.figure(figsize=(5,5))
sns.scatterplot(x=df['Insulin'],y=df['Glucose'],data=df,hue=df['Outcome'])
```
